```python
import pandas as pd
import numpy as np


grades = {'Roll No' : [i for i in range(1,51)],
          'Science' : np.random.randint(25,95,size=50),
          'Maths' : np.random.randint(25,95,size=50),
          'English' : np.random.randint(25,95,size=50)}

grades = pd.DataFrame(grades)

grades.head()
```

```python
def mapper(grades):
    result = []

    for _,row in grades.iterrows():
        roll_no = row['Roll No']
        for subject in ['Science','Maths','English']:
            result.append((roll_no,row[subject]))

    return result
```

```python
mapped_output = mapper(grades)
mapped_output
```

```python
def shuffle_sort(mapped_output):
    result = {}

    for roll,marks in mapped_output:
        if roll in result:
            result[roll].append(marks)
        else:
            result[roll] = [marks]

    return result
```

```python
shuffle_sort_output = shuffle_sort(mapped_output)
shuffle_sort_output
```

```python
def reducer(shuffle_sort_output):
    result = {'Roll No': [],
              'Total': [],
              'Grade': []}

    for roll,marks in shuffle_sort_output.items():
        result['Roll No'].append(roll)

        result['Total'].append(sum(marks))

        if result['Total'][-1] > 210:
            result['Grade'].append('A')
        elif 150 < result['Total'][-1] < 210:
            result['Grade'].append('B')
        else:
            result['Grade'].append('C')

    result = pd.DataFrame(result)

    return result
```

```python
reducer_output = reducer(shuffle_sort_output)
reducer_output
```
